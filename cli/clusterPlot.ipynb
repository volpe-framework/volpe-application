```python
import numpy as np
import pandas as pd
import sklearn.decomposition as decomp
import sklearn.cluster as cluster
```

```python
df = pd.read_csv("../cli/F12_summary.csv")
df.head()
```

```python
def string_to_array(s):
    slist = s[1:-1].split(' ')
    filtered_list = []
    for item in slist:
        stripped_item = item.strip()
        if stripped_item:
            filtered_list.append(float(stripped_item))
    return np.array(filtered_list)
```

```python
genotype_strings = df.genotype
genotype = np.array([ string_to_array(s) for s in genotype_strings ])
genotype.shape
```

```python
g_normalized = (genotype - -100) / (100 - -100)
g_rounded = np.floor(g_normalized*N_PARTITIONS)
g_rounded.shape
```

```python
g_strings = [ '*'.join(x.astype(int).astype(str)) for x in g_rounded ]
df.genotype_hash = g_strings
df[:5]
```
