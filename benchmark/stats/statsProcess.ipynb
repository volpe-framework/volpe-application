```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def get_deap_stats(problem):
    files = [ f"deap_{problem}_c5_stats.txt", f"deap_{problem}_c3_stats.txt" ]
    dfs = [pd.read_csv(f) for f in files]
    cpu_agg = np.mean(dfs[0]['CPU_Usage_Percent'])*5/8 + np.mean(dfs[1]['CPU_Usage_Percent'])*3/8
    cpu_std = np.std(dfs[0]['CPU_Usage_Percent'])*5/8 + np.std(dfs[1]['CPU_Usage_Percent'])*3/8
    mem_agg = np.max(dfs[0]['Memory_Usage_GB']) + np.max(dfs[1]['Memory_Usage_GB'])
    mem_std = np.std(dfs[0]['Memory_Usage_GB'])*5/8 + np.std(dfs[1]['Memory_Usage_GB'])*3/8

    return cpu_agg, cpu_std, mem_agg, mem_std
```

```python
get_deap_stats('f7')
```

```python
def get_prop_stats(problem):
    files = [ f"prop_{problem}_c5_stats.txt", f"prop_{problem}_c3_stats.txt" ]
    dfs = [pd.read_csv(f) for f in files]
    cpu_agg = np.mean(dfs[0]['CPU_Usage_Percent'])*5/8 + np.mean(dfs[1]['CPU_Usage_Percent'])*3/8
    cpu_std = np.std(dfs[0]['CPU_Usage_Percent'])*5/8 + np.std(dfs[1]['CPU_Usage_Percent'])*3/8
    mem_agg = np.max(dfs[0]['Memory_Usage_GB']) + np.max(dfs[1]['Memory_Usage_GB'])
    mem_std = np.std(dfs[0]['Memory_Usage_GB'])*5/8 + np.std(dfs[1]['Memory_Usage_GB'])*3/8

    return cpu_agg, cpu_std, mem_agg, mem_std
```

```python
get_prop_stats('f7')
```

```python
def get_seq_stats(problem):
    file = f"seq_{problem}_stats.txt"
    df = pd.read_csv(file)
    cpu_agg = np.mean(df['CPU_Usage_Percent'])
    cpu_std = np.std(df['CPU_Usage_Percent'])
    mem_agg = np.max(df['Memory_Usage_GB'])
    mem_std = np.std(df['Memory_Usage_GB'])
    return cpu_agg, cpu_std, mem_agg, mem_std
```

```python
get_seq_stats('f7')
```

```python
def get_dist_stats(problem):
    file = f"dist_{problem}_stats.txt"
    df = pd.read_csv(file)
    cpu_agg = np.mean(df['CPU_Usage_Percent'])
    cpu_std = np.std(df['CPU_Usage_Percent'])
    mem_agg = np.max(df['Memory_Usage_GB'])
    mem_std = np.std(df['Memory_Usage_GB'])
    return cpu_agg, cpu_std, mem_agg, mem_std
```

```python
get_dist_stats('f7')
```

```python
def make_line(problem):
    deap_stats = ' & '.join([ f'{stat:.2f}' for stat in get_deap_stats(problem)] )
    prop_stats = ' & '.join([ f'{stat:.2f}' for stat in get_prop_stats(problem)] )
    seq_stats = ' & '.join([ f'{stat:.2f}' for stat in get_seq_stats(problem)] )
    dist_stats = ' & '.join([ f'{stat:.2f}' for stat in get_dist_stats(problem)] )

    return f"{problem.upper()} & {seq_stats} & {dist_stats} & {deap_stats} & {prop_stats} \\\\"
```

```python
for problem in ['f7', 'f11', 'f12']:
    print(make_line(problem))
```
